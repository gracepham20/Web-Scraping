```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#import important Data Science packages

import requests
#import library

!pip install lxml beautifulsoup4 html5lib
#install additional libraries used by Pandas's function read_html
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_past_and_estimated_future_population'
resp = requests.get(url)
soup = BeautifulSoup(resp.content, 'html.parser')
tables = soup.find() #finding tables in HTML file

table1 = pd.read_html(str(tables))[0] #getting first table
table2 = pd.read_html(str(tables))[1] #getting second table
table3 = pd.read_html(str(tables))[2] #getting third table
```

```python
df = pd.concat([table1, table2.iloc[:,1:],table3.iloc[:,1:]], axis =1) #concatenating 3 tables into dataframe "df"
df
```

```python
df = df[:-1]
df
#excluding last row about world population data
```

```python
df.columns = ['Country','1950', '1955', 'AAG1955', '1960', 'AAG1960', '1965', 'AAG1965', '1970', 'AAG1970',
       '1975', 'AAG1975', '1980', 'AAG1980', '1985', 'AAG1985', '1990', 'AAG1990', '1995', 'AAG1995',
       '2000', 'AAG2000', '2005', 'AAG2005', '2010', 'AAG2010', '2015', 'AAG2015', '2020', 'AAG2020',
       '2025', 'AAG2025', '2030', 'AAG2030', '2035', 'AAG2035', '2040', 'AAG2040', '2045',
       'AAG2045', '2050', 'AAG2050']
#rename column names in df

df= df.set_index('Country')
#set the index column of dataframe "df" to column "Country"

df.head()
```

```python
p_countries = df.sort_values(by=['1950'], ascending=False).head(10)
#dataframe "p_countries" containing data of top 10 most populous countries in 1950
```

```python
p_countries1 = p_countries[['1950', '1955', '1960','1965','1970','1975','1980','1985','1990','1995','2000','2005','2010','2015','2020','2025','2030','2035','2040','2045','2050']]

p_countries1.stack().swaplevel().unstack()
#stacking "p_countries1"

years = p_countries1.T.index[:]

#plotting population data from 1950-2050 (in 5-year interval) for top 10 most populous countries in 1950
ax = p_countries1.T.plot.line(figsize = (12,8), marker='o')
ax.set_xticks(range(22))
ax.set_xticklabels(years, rotation = 45)
plt.ylabel('Population in Thousands')
plt.xlabel('Year')
plt.title('Population in Top 10 Most Populous Countries in 1950 (1950-2050)')
plt.tight_layout()
```

```python
p_countries2 = p_countries[['AAG1955','AAG1960', 'AAG1965',
       'AAG1970', 'AAG1975', 'AAG1980','AAG1985',
       'AAG1990', 'AAG1995','AAG2000', 'AAG2005',
       'AAG2010', 'AAG2015', 'AAG2020', 'AAG2025',
       'AAG2030','AAG2035', 'AAG2040',
       'AAG2045', 'AAG2050']]

years2 = p_countries1.T.index[-20:]

#plotting the values of average annual growth from 1955 to 2050 in 5-year interval
ax2 = p_countries2.T.plot.line(figsize = (12,8), marker='o')
ax2.set_xticks(range(22))
ax2.set_xticklabels(years2, rotation = 45)
plt.ylabel('Average Annual Growth in Previous 5 Years (%)')
plt.xlabel('Year')
plt.title('Average Annual Growth in Top 10 Most Populous Countries in 1950 (1955-2050)')
plt.tight_layout()
```
